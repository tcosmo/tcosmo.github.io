```python
data = np.random.rand(1000,2)

my_som = som(40,40,2)
inter = my_som.train(data,sigma0=20,L0=0.1,lam=30,stock_inter=True,pp=False)
```

```python
def get_x_y(som):
    x = []
    y = []
    for a in som.som:
        for b in a:
            x.append(b[0])
            y.append(b[1])
    return x,y
```

```python
x,y = get_x_y(my_som)
```

```python
#Reference: https://stackoverflow.com/questions/8487893/generate-all-the-points-on-the-circumference-of-a-circle
def PointsInCircum(r,n=1000):
    return [(math.cos(2*np.pi/n*x)*r+0.5,math.sin(2*np.pi/n*x)*r+0.5) for x in range(0,n+1)]
```
